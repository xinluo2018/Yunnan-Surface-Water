#### Calculate the standard deviation of the time-series heights.

```python
mean_chenghai, std_chenghai = np.mean(list(heights_lake_chenghai.values())), np.std(list(heights_lake_chenghai.values()))
mean_dianchi, std_dianchi = np.mean(list(heights_lake_dianchi.values())), np.std(list(heights_lake_dianchi.values()))
mean_erhai, std_erhai = np.mean(list(heights_lake_erhai.values())), np.std(list(heights_lake_erhai.values()))
mean_fuxian, std_fuxian = np.mean(list(heights_lake_fuxian.values())), np.std(list(heights_lake_fuxian.values()))
mean_lugu, std_lugu = np.mean(list(heights_lake_lugu.values())),np.std(list(heights_lake_lugu.values()))  
mean_qilu, std_qilu = np.mean(list(heights_lake_qilu.values())), np.std(list(heights_lake_qilu.values()))
mean_xingyun, std_xingyun = np.mean(list(heights_lake_xingyun.values())), np.std(list(heights_lake_xingyun.values()))
mean_yangzonghai, std_yangzonghai = np.mean(list(heights_lake_yangzonghai.values())), np.std(list(heights_lake_yangzonghai.values()))
mean_yilong, std_yilong = np.mean(list(heights_lake_yilong.values())), np.std(list(heights_lake_yilong.values()))
print('mean:', mean_chenghai, mean_dianchi, mean_erhai, mean_fuxian, mean_lugu, mean_qilu, mean_xingyun, mean_yangzonghai, mean_yilong)
print('std:', std_chenghai, std_dianchi, std_erhai, std_fuxian, std_lugu, std_qilu, std_xingyun, std_yangzonghai, std_yilong)
```

```python
max_range_chenghai = np.max(list(heights_lake_chenghai.values())) - np.min(list(heights_lake_chenghai.values()))
max_range_dianchi = np.max(list(heights_lake_dianchi.values())) - np.min(list(heights_lake_dianchi.values()))
max_range_erhai = np.max(list(heights_lake_erhai.values())) - np.min(list(heights_lake_erhai.values()))
max_range_fuxian = np.max(list(heights_lake_fuxian.values())) - np.min(list(heights_lake_fuxian.values()))
max_range_lugu = np.max(list(heights_lake_lugu.values())) - np.min(list(heights_lake_lugu.values()))
max_range_qilu = np.max(list(heights_lake_qilu.values())) - np.min(list(heights_lake_qilu.values()))
max_range_xingyun = np.max(list(heights_lake_xingyun.values())) - np.min(list(heights_lake_xingyun.values()))
max_range_yangzonghai = np.max(list(heights_lake_yangzonghai.values())) - np.min(list(heights_lake_yangzonghai.values()))
max_range_yilong = np.max(list(heights_lake_yilong.values())) - np.min(list(heights_lake_yilong.values()))
print('max_range:', max_range_chenghai, max_range_dianchi, max_range_erhai, max_range_fuxian, max_range_lugu, max_range_qilu, max_range_xingyun, max_range_yangzonghai, max_range_yilong)
mean_max_range = np.mean([max_range_chenghai, max_range_dianchi, max_range_erhai, max_range_fuxian, max_range_lugu, max_range_qilu, max_range_xingyun, max_range_yangzonghai, max_range_yilong])
print('mean_max_range:', mean_max_range)
```

#### Get the date of minimal lake level and maximal lake level.

```python
def get_date_min_max(heights_lake):
    id_min = np.argmin(list(heights_lake.values()))
    id_max = np.argmax(list(heights_lake.values()))
    date_min = list(heights_lake.keys())[id_min]
    date_max = list(heights_lake.keys())[id_max]
    return date_min, date_max
```

```python
date_min_chenghai,date_max_chenghai = get_date_min_max(heights_lake_chenghai)
date_min_dianchi,date_max_dianchi = get_date_min_max(heights_lake_dianchi)
date_min_erhai,date_max_erhai = get_date_min_max(heights_lake_erhai)
date_min_fuxian,date_max_fuxian = get_date_min_max(heights_lake_fuxian)
date_min_lugu,date_max_lugu = get_date_min_max(heights_lake_lugu)
date_min_qilu,date_max_qilu = get_date_min_max(heights_lake_qilu)
date_min_xingyun,date_max_xingyun = get_date_min_max(heights_lake_xingyun)
date_min_yangzonghai,date_max_yangzonghai = get_date_min_max(heights_lake_yangzonghai)
date_min_yilong,date_max_yilong = get_date_min_max(heights_lake_yilong)
print('date_min:', date_min_chenghai,date_min_dianchi,date_min_erhai,date_min_fuxian,date_min_lugu,date_min_qilu,date_min_xingyun,date_min_yangzonghai,date_min_yilong)
print('date_max:', date_max_chenghai,date_max_dianchi,date_max_erhai,date_max_fuxian,date_max_lugu,date_max_qilu,date_max_xingyun,date_max_yangzonghai,date_max_yilong)
```

```python
def plot_lake_heights(heights_lake, date_ticks, ax, h_range=None, markersize=6, h_color='blue', h_label='Lake level'):
  dates_decimal = [date_to_decimal(date+'T00:00:00') for date in date_ticks]
  ax.plot(heights_lake.keys(), heights_lake.values(), '--.', 
    color=h_color, linewidth=0.6, markersize=markersize, label=h_label)
  ax.set_xticks(dates_decimal); ax.set_xticklabels(date_ticks, rotation=0)
  ax.legend()
  if h_range is not None:
    ax.set_ylim([h_range[0], h_range[1]]); 
    ax.set_yticks([h_range[0]+0.1+i*0.4 for i in range(0, 4)]);
```

### Flucations for lakes 

```python
fluctions_lake_chenghai, fluctions_lake_dianchi, fluctions_lake_erhai = {}, {}, {}
fluctions_lake_fuxian, fluctions_lake_lugu, fluctions_lake_qilu = {}, {}, {}
fluctions_lake_xingyun, fluctions_lake_yangzonghai, fluctions_lake_yilong = {}, {}, {}
for key in heights_lake_chenghai.keys():
    fluctions_lake_chenghai[key] = heights_lake_chenghai[key] - mean_chenghai
for key in heights_lake_dianchi.keys():
    fluctions_lake_dianchi[key] = heights_lake_dianchi[key] - mean_dianchi
for key in heights_lake_erhai.keys():
    fluctions_lake_erhai[key] = heights_lake_erhai[key] - mean_erhai
for key in heights_lake_fuxian.keys():  
    fluctions_lake_fuxian[key] = heights_lake_fuxian[key] - mean_fuxian
for key in heights_lake_lugu.keys():
    fluctions_lake_lugu[key] = heights_lake_lugu[key] - mean_lugu
for key in heights_lake_qilu.keys():
    fluctions_lake_qilu[key] = heights_lake_qilu[key] - mean_qilu
for key in heights_lake_xingyun.keys():
    fluctions_lake_xingyun[key] = heights_lake_xingyun[key] - mean_xingyun
for key in heights_lake_yangzonghai.keys():
    fluctions_lake_yangzonghai[key] = heights_lake_yangzonghai[key] - mean_yangzonghai
for key in heights_lake_yilong.keys():  
    fluctions_lake_yilong[key] = heights_lake_yilong[key] - mean_yilong
```

```python
dates_ticks = ['2023-08-01','2023-12-01', '2024-04-01', '2024-08-01', '2024-12-01']
fig, ax = plt.subplots(1,1,figsize=(10, 4))
plot_lake_heights(heights_lake=fluctions_lake_chenghai, date_ticks=dates_ticks, markersize=8,
                  h_color='blue', h_label='Chenghai Lake', ax=ax)
plot_lake_heights(heights_lake=fluctions_lake_dianchi, date_ticks=dates_ticks, markersize=8,
                  h_color='orange', h_label='Dianchi Lake', ax=ax)
plot_lake_heights(heights_lake=fluctions_lake_erhai, date_ticks=dates_ticks, markersize=8,
                  h_color='green', h_label='Erhai Lake', ax=ax)
plot_lake_heights(heights_lake=fluctions_lake_fuxian, date_ticks=dates_ticks, markersize=8,
                  h_color='red', h_label='Fuxian Lake', ax=ax)
plot_lake_heights(heights_lake=fluctions_lake_qilu, date_ticks=dates_ticks, markersize=8,
                  h_color='purple', h_label='Lugu Lake', ax=ax)
plot_lake_heights(heights_lake=fluctions_lake_qilu, date_ticks=dates_ticks, markersize=8,
                  h_color='brown', h_label='Qilu Lake', ax=ax)
plot_lake_heights(heights_lake=fluctions_lake_xingyun, date_ticks=dates_ticks, markersize=8,
                  h_color='pink', h_label='Xingyun Lake', ax=ax)
plot_lake_heights(heights_lake=fluctions_lake_yangzonghai, date_ticks=dates_ticks, markersize=8,
                  h_color='gray', h_label='Yangzonghai Lake', ax=ax)
plot_lake_heights(heights_lake=fluctions_lake_yilong, date_ticks=dates_ticks, markersize=8,
                  h_color='black', h_label='Yilong Lake', ax=ax)

ax.set_xlabel('Date', fontsize=13)
ax.set_ylabel('Fluctuations (m)', fontsize=13)
ax.set_title('Lake level fluctuations (m)', fontsize=14)
ax.legend(ncol=3, loc='upper center', fontsize=9, bbox_to_anchor=(0.43, 1.01), frameon=False)
ax.tick_params(axis='x', labelsize=13)
ax.tick_params(axis='y', labelsize=13)
ax.set_ylim([-0.8, 0.8])
ax.set_yticks([-0.6+i*0.3 for i in range(0,5)]);

season_ticks = (np.array([2023.5, 2023.75, 2024.0, 2024.25, 2024.5, 2024.75, 2025.0])-ax.get_xlim()[0]) / (ax.get_xlim()[1]-ax.get_xlim()[0]) 
colors = ["#90EE90", "#E0FFFF", "#8B0000", "#000000"]
ax.axhspan(ymin=-1, ymax=1, xmin=0, xmax=season_ticks[0], color=colors[1], alpha=0.15)
ax.axhspan(ymin=-1, ymax=1, xmin=season_ticks[0], xmax=season_ticks[1], color=colors[2], alpha=0.15)
ax.axhspan(ymin=-1, ymax=1, xmin=season_ticks[1], xmax=season_ticks[2], color=colors[3], alpha=0.15)
ax.axhspan(ymin=-1, ymax=1, xmin=season_ticks[2], xmax=season_ticks[3], color=colors[0], alpha=0.15)
ax.axhspan(ymin=-1, ymax=1, xmin=season_ticks[3], xmax=season_ticks[4], color=colors[1], alpha=0.15)
ax.axhspan(ymin=-1, ymax=1, xmin=season_ticks[4], xmax=season_ticks[5], color=colors[2], alpha=0.15)
ax.axhspan(ymin=-1, ymax=1, xmin=season_ticks[5], xmax=season_ticks[6], color=colors[3], alpha=0.15)
ax.axhspan(ymin=-1, ymax=1, xmin=season_ticks[6], xmax=1, color=colors[0], alpha=0.15)
```
