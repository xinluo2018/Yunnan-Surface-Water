```python
import pickle 
import numpy as np
import matplotlib.pyplot as plt
from scripts.date_transform import date_to_decimal
```

```python
path_lake_level_chenghai = 'data/swot_l2/pixc/chenghai-lake/chenghai_heights_lake.pkl'
path_lake_level_dianchi = 'data/swot_l2/pixc/dianchi-lake/dianchi_heights_lake.pkl'
path_lake_level_erhai = 'data/swot_l2/pixc/erhai-lake/erhai_heights_lake.pkl'
path_lake_level_fuxian = 'data/swot_l2/pixc/fuxian-lake/fuxian_heights_lake.pkl'
path_lake_level_lugu = 'data/swot_l2/pixc/lugu-lake/lugu_heights_lake.pkl'
path_lake_level_qilu = 'data/swot_l2/pixc/qilu-lake/qilu_heights_lake.pkl'
path_lake_level_xingyun = 'data/swot_l2/pixc/xingyun-lake/xingyun_heights_lake.pkl'
path_lake_level_yangzonghai = 'data/swot_l2/pixc/yangzonghai-lake/yangzonghai_heights_lake.pkl'
path_lake_level_yilong = 'data/swot_l2/pixc/yilong-lake/yilong_heights_lake.pkl'
```

#### read time-series lake levels

```python
with open(path_lake_level_chenghai, 'rb') as f: 
    heights_lake_chenghai = pickle.load(f)
with open(path_lake_level_dianchi, 'rb') as f:
    heights_lake_dianchi = pickle.load(f)
with open(path_lake_level_erhai, 'rb') as f:
    heights_lake_erhai = pickle.load(f)
with open(path_lake_level_fuxian, 'rb') as f:
    heights_lake_fuxian = pickle.load(f)
with open(path_lake_level_lugu, 'rb') as f:
    heights_lake_lugu = pickle.load(f)
with open(path_lake_level_qilu, 'rb') as f:
    heights_lake_qilu = pickle.load(f)
with open(path_lake_level_xingyun, 'rb') as f:
    heights_lake_xingyun = pickle.load(f)
with open(path_lake_level_yangzonghai, 'rb') as f:
    heights_lake_yangzonghai = pickle.load(f)
with open(path_lake_level_yilong, 'rb') as f:
    heights_lake_yilong = pickle.load(f)
```

#### Calculate the standard deviation of the time-series heights.

```python
std_chenghai = np.std(list(heights_lake_chenghai.values()))
std_dianchi = np.std(list(heights_lake_dianchi.values()))
std_erhai = np.std(list(heights_lake_erhai.values()))
std_fuxian = np.std(list(heights_lake_fuxian.values()))
std_lugu = np.std(list(heights_lake_lugu.values())) 
std_qilu = np.std(list(heights_lake_qilu.values()))
std_xingyun = np.std(list(heights_lake_xingyun.values()))
std_yangzonghai = np.std(list(heights_lake_yangzonghai.values()))
std_yilong = np.std(list(heights_lake_yilong.values()))
print('std:', std_chenghai, std_dianchi, std_erhai, std_fuxian, std_lugu, std_qilu, std_xingyun, std_yangzonghai, std_yilong)
```

```python
max_range_chenghai = np.max(list(heights_lake_chenghai.values())) - np.min(list(heights_lake_chenghai.values()))
max_range_dianchi = np.max(list(heights_lake_dianchi.values())) - np.min(list(heights_lake_dianchi.values()))
max_range_erhai = np.max(list(heights_lake_erhai.values())) - np.min(list(heights_lake_erhai.values()))
max_range_fuxian = np.max(list(heights_lake_fuxian.values())) - np.min(list(heights_lake_fuxian.values()))
max_range_lugu = np.max(list(heights_lake_lugu.values())) - np.min(list(heights_lake_lugu.values()))
max_range_qilu = np.max(list(heights_lake_qilu.values())) - np.min(list(heights_lake_qilu.values()))
max_range_xingyun = np.max(list(heights_lake_xingyun.values())) - np.min(list(heights_lake_xingyun.values()))
max_range_yangzonghai = np.max(list(heights_lake_yangzonghai.values())) - np.min(list(heights_lake_yangzonghai.values()))
max_range_yilong = np.max(list(heights_lake_yilong.values())) - np.min(list(heights_lake_yilong.values()))
print('max_range:', max_range_chenghai, max_range_dianchi, max_range_erhai, max_range_fuxian, max_range_lugu, max_range_qilu, max_range_xingyun, max_range_yangzonghai, max_range_yilong)
mean_max_range = np.mean([max_range_chenghai, max_range_dianchi, max_range_erhai, max_range_fuxian, max_range_lugu, max_range_qilu, max_range_xingyun, max_range_yangzonghai, max_range_yilong])
print('mean_max_range:', mean_max_range)
```

#### Get the date of minimal lake level and maximal lake level.

```python
def get_date_min_max(heights_lake):
    id_min = np.argmin(list(heights_lake.values()))
    id_max = np.argmax(list(heights_lake.values()))
    date_min = list(heights_lake.keys())[id_min]
    date_max = list(heights_lake.keys())[id_max]
    return date_min, date_max
```

```python
date_min_chenghai,date_max_chenghai = get_date_min_max(heights_lake_chenghai)
date_min_dianchi,date_max_dianchi = get_date_min_max(heights_lake_dianchi)
date_min_erhai,date_max_erhai = get_date_min_max(heights_lake_erhai)
date_min_fuxian,date_max_fuxian = get_date_min_max(heights_lake_fuxian)
date_min_lugu,date_max_lugu = get_date_min_max(heights_lake_lugu)
date_min_qilu,date_max_qilu = get_date_min_max(heights_lake_qilu)
date_min_xingyun,date_max_xingyun = get_date_min_max(heights_lake_xingyun)
date_min_yangzonghai,date_max_yangzonghai = get_date_min_max(heights_lake_yangzonghai)
date_min_yilong,date_max_yilong = get_date_min_max(heights_lake_yilong)
print('date_min:', date_min_chenghai,date_min_dianchi,date_min_erhai,date_min_fuxian,date_min_lugu,date_min_qilu,date_min_xingyun,date_min_yangzonghai,date_min_yilong)
print('date_max:', date_max_chenghai,date_max_dianchi,date_max_erhai,date_max_fuxian,date_max_lugu,date_max_qilu,date_max_xingyun,date_max_yangzonghai,date_max_yilong)
```

```python
dates_ticks = ['2023-09-01','2024-01-01', '2024-05-01', '2024-09-01', '2025-01-01']
fig, ax = plt.subplots(1,2,figsize=(12, 3))
## Chenghai
plot_lake_heights(heights_lake=heights_lake_chenghai, 
                  date_ticks=dates_ticks, h_range=[1495.95, 1497.45], ax=ax[0])
ax[0].set_title('Lake level (m) in Chenghai Lake', fontsize=10)
## Dianchi
plot_lake_heights(heights_lake=heights_lake_dianchi, 
                  date_ticks=dates_ticks, h_range=[1887., 1888.5], ax=ax[1])
ax[1].set_title('Lake level (m) in Dianchi Lake', fontsize=10)
```
